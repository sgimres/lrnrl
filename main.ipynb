```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([7,6,8,5,6,9])
y = np.array([12,8,12,10,11,13])
```

```python
# m = number of tarning samples
m = x.shape[0]

w, b  =0,0
f_wb_x = w*x + b
f_wb_x
```

```python
plt.scatter(x, y, marker="x")
plt.plot(x, f_wb_x)
```

```python
def calculate_cost(x, y, w, b):
    m = x.shape[0]
    cost = 0
    for i in range(m):
        j_wb = (w * x[i] + b - y[i])**2
        cost += j_wb
    cost /= 2*m
    return cost 
```

```python
calculate_cost(x, y, w, b)
```

```python
def calculate_gradient(x, y, w, b):
    m = x.shape[0]
    dj_dw = 0
    dj_db = 0 
    for i in range(m):
        dj_dw_i = (w*x[i] +b -y[i])*x[i]
        dj_db_i = (w*x[i] +b -y[i])
        dj_dw += dj_dw_i
        dj_db += dj_db_i
    dj_dw /= m
    dj_db /= m
    return dj_dw, dj_db
```

```python
calculate_gradient(x, y , w, b)
```

```python
import math
def gradient_descent(x, y, w, b, iter, cost_function, gradient_function, alpha ):
    m = x.shape[0]
    cost_hist = []
    w_hist = []
    for i in range(iter):
        dj_dw , dj_db = gradient_function(x,y,w,b)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        if i < 1000000 : 
            cost = cost_function(x, y ,w,b)
            cost_hist.append(cost)
        
        if i% math.ceil(iter/10) == 0 or i == (iter-1):
            w_hist.append(w)
            print(f"Iteration {i:4}: Cost {float(cost_hist[-1]):8.2f}   ")
    return w, b
```

```python
n_w, n_b = gradient_descent(x,y,w,b, 10000, calculate_cost, calculate_gradient, 0.01)
```

```python
print(n_w, n_b)
```

```python
plt.scatter(x,y, marker='x')
line = (x * n_w) + n_b
plt.plot(x, line)
```
